```python
import pandas as pd
import re

# Extracted text sample from the PDF
text = """
Nome Inscrição Língua Portuguesa Raciocínio Lógico-Matemático Noções de Direito Administrativo Noções de Direito Constitucional Legislação Específica/Regim. Interno Noções de Ciências Pólíticas Realidade sócio-político-historica do RS Conhecimentos Específicos Políticas Públicas Total da Nota
Abigail dos Reis Cruz 87210874456-2 - - - - - - - -
Abigail Ferraz Borges 87210886942-9 18,00 8,00 7,00 1,00 3,00 24,00 2,00 63,00
Abilio Angelo da Costa Neto 87210881805-7 16,00 16,00 5,00 3,00 2,00 26,00 0,00 68,00
Abner William Silva de Mattos 87210869820-3 8,00 8,00 4,00 1,00 3,00 22,00 0,00 46,00
"""

# Extract the table headers
headers = [
    "Nome", "Inscrição", "Língua Portuguesa", "Raciocínio Lógico-Matemático",
    "Noções de Direito Administrativo", "Noções de Direito Constitucional",
    "Legislação Específica/Regim. Interno", "Noções de Ciências Pólíticas",
    "Realidade sócio-político-historica do RS", "Conhecimentos Específicos",
    "Políticas Públicas", "Total da Nota"
]

# Using regular expressions to parse the data
pattern = re.compile(r'(\d{11}-\d)\s+([\w\s]+)\s+([\d,]+|-)\s+([\d,]+|-)\s+([\d,]+|-)\s+([\d,]+|-)\s+([\d,]+|-)\s+([\d,]+|-)\s+([\d,]+|-)\s+([\d,]+|-)\s+([\d,]+|-)\s+([\d,]+|-)\s+([\d,]+|-)\s+(\d{1,3},\d{2})')
matches = pattern.findall(text)

# Create a list of dictionaries from the extracted data
data = []
for match in matches:
    entry = {
        'Inscrição': match[0].strip(),
        'Nome': match[1].strip(),
        'Língua Portuguesa': match[2].replace(',', '.'),
        'Raciocínio Lógico-Matemático': match[3].replace(',', '.'),
        'Noções de Direito Administrativo': match[4].replace(',', '.'),
        'Noções de Direito Constitucional': match[5].replace(',', '.'),
        'Legislação Específica/Regim. Interno': match[6].replace(',', '.'),
        'Noções de Ciências Pólíticas': match[7].replace(',', '.'),
        'Realidade sócio-político-historica do RS': match[8].replace(',', '.'),
        'Conhecimentos Específicos': match[9].replace(',', '.'),
        'Políticas Públicas': match[10].replace(',', '.'),
        'Total da Nota': match[13].replace(',', '.'),
    }
    data.append(entry)

# Convert the list of dictionaries into a DataFrame
df = pd.DataFrame(data)

# Convert the 'Total da Nota' column to float for sorting
df['Total da Nota'] = df['Total da Nota'].astype(float)

# Sort the DataFrame by 'Total da Nota' in ascending order
df_sorted = df.sort_values(by='Total da Nota')

# Save the sorted DataFrame to a CSV file
output_csv_path = '/mnt/data/Notas_Preliminares_TO_872_66722b67962e4_sorted_ascending.csv'
df_sorted.to_csv(output_csv_path, index=False)

output_csv_path
```
